10. Log File Analyzer

Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,
analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface.

```python

log_data = [
    "[2026-06-19 10:15:30] [INFO] [User:admin] User logged in successfully.",
    "[2026-06-19 10:16:12] [ERROR] [User:john_doe] Database connection failed.",
    "[2026-06-19 10:18:45] [WARNING] [User:alice] Disk space running low (85%).",
    "[2026-06-19 10:20:01] [INFO] [User:john_doe] File uploaded successfully.",
    "[2026-06-19 10:22:15] [ERROR] [User:admin] Failed to update settings.",
    "[2026-06-19 10:25:30] [INFO] [User:bob] User logged in successfully."
]

def display_menu():
    print("\n" + "="*35)
    print("        LOG FILE ANALYZER")
    print("="*35)
    print("1. View All Logs")
    print("2. Count Severity Levels (Errors/Warnings/Info)")
    print("3. Analyze User Activity")
    print("4. Filter Logs by Severity")
    print("5. Exit")
    print("="*35)

# Main program loop
while True:
    display_menu()
    choice = input("Enter your choice (1-5): ").strip()

    if choice == '1':
        print("\n--- All Log Entries ---")
        for log in log_data:
            print(log)

    elif choice == '2':
        print("\n--- Severity Counts ---")
        # Dictionary to store aggregation counts
        severity_counts = {}

        for log in log_data:
            # String splitting to extract the severity tag
            # Example: splitting by ']' gives ['[2026-06-19 10:15:30', ' [INFO', ' [User:admin', ' User...']
            parts = log.split(']')
            if len(parts) > 1:

                severity = parts[1].replace('[', '').strip()


                if severity in severity_counts:
                    severity_counts[severity] += 1
                else:
                    severity_counts[severity] = 1

        for level, count in severity_counts.items():
            print(f"{level}: {count} time(s)")

    elif choice == '3':
        print("\n User Activity Summary")
        user_counts = {}

        for log in log_data:
            # Pattern matching logic using simple string split/find
            if "[User:" in log:
                # Find start and end indices of username
                start = log.find("[User:") + 6
                end = log.find("]", start)
                username = log[start:end]

                # Dictionary-based aggregation
                if username in user_counts:
                    user_counts[username] += 1
                else:
                    user_counts[username] = 1

        for user, count in user_counts.items():
            print(f"User '{user}' performed {count} action(s)")

    elif choice == '4':
        severity_filter = input("Enter severity to filter (INFO, WARNING, ERROR): ").upper().strip()
        print(f"\n--- Logs matching [{severity_filter}] ---")

        found = False
        for log in log_data:
            # Match the exact pattern block
            if f"[{severity_filter}]" in log:
                print(log)
                found = True
        if not found:
            print("No matching logs found.")

    elif choice == '5':
        print("\nExiting Log File Analyzer. Goodbye!")
        break

    else:
        print("\nInvalid choice! Please enter a number between 1 and 5.")
```
